```python
def convert_units(value, conversion_type):

    if conversion_type == 1:  # meters to feet
        return value * 3.28084
    elif conversion_type == 2:  # feet to meters
        return value / 3.28084
    elif conversion_type == 3:  # kg to pounds
        return value * 2.20462
    elif conversion_type == 4:  # pounds to kg
        return value / 2.20462
    elif conversion_type == 5:  # liters to gallons
        return value * 0.264172
    elif conversion_type == 6:  # gallons to liters
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("\nChoose Conversion Type:")
    print("1. Length: meters to feet")
    print("2. Length: feet to meters")
    print("3. Weight: kilograms to pounds")
    print("4. Weight: pounds to kilograms")
    print("5. Volume: liters to gallons")
    print("6. Volume: gallons to liters")

    try:
        choice = int(input("\nEnter your choice (1-6): "))
        value = float(input("Enter the value to convert: "))

        result = convert_units(value, choice)

        if result is None:
            print("Unsupported conversion type.")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


if __name__ == "__main__":
    main()
```

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of all numbers in the list.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average value of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Finds the maximum number in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The largest number in the list.
    """
    return max(numbers)


def find_min(numbers):
    """
    Finds the minimum number in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The smallest number in the list.
    """
    return min(numbers)


def main():
    print("Mathematical Operations on a List")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    try:
        choice = int(input("Choose an operation (1-4): "))

        user_input = input("Enter numbers separated by spaces: ")
        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            raise ValueError("The list is empty.")

        if choice == 1:
            result = find_sum(numbers)
            print("Sum:", result)

        elif choice == 2:
            result = find_average(numbers)
            print("Average:", result)

        elif choice == 3:
            result = find_max(numbers)
            print("Maximum:", result)

        elif choice == 4:
            result = find_min(numbers)
            print("Minimum:", result)

        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Invalid input! Please enter valid numbers.")


if __name__ == "__main__":
    main()
```

Nested list

```python
def flatten(lst):

    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(item)  # add elements of sublist
        else:
            flat_list.append(item)  # add single element
    return flat_list

nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)
print("Flattened List:", result)
```

Accessing Nested List Elements:

```python
def access_nested_element(lst, indices):
    element = lst
    for idx in indices:
        element = element[idx]  # go deeper into the nested list
    return element

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)
print("Output:", result)
```

Sum of All Elements in a Nested List:

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, including sublists at any depth.

    Parameters:
    lst (list): A nested list of numbers.

    Returns:
    int/float: Sum of all numbers in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call for sublist
        else:
            total += item  # add number directly
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)
print("Sum of all elements:", result)
```

Remove Specific Element from a Nested List:

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from a nested list.

    Parameters:
    lst (list): A nested list of elements.
    elem: The element to remove from the list.

    Returns:
    list: A new nested list with all occurrences of elem removed.
    """
    new_list = []
    for item in lst:
        if isinstance(item, list):
            # Recursively remove element from sublist
            new_list.append(remove_element(item, elem))
        else:
            if item != elem:
                new_list.append(item)
    return new_list

nested_list = [[1, 2], [3, 2], [4, 5]]
result = remove_element(nested_list, 2)
print("Modified List:", result)
```

Find the Maximum Element in a Nested List:

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list of numbers.

    Returns:
    int/float: The maximum number in the nested list.
    """
    max_value = float('-inf')  # start with smallest possible number
    for item in lst:
        if isinstance(item, list):
            # recursive call for sublist
            sub_max = find_max(item)
            if sub_max > max_value:
                max_value = sub_max
        else:
            if item > max_value:
                max_value = item
    return max_value

nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)
print("Maximum Element:", result)
```

Count Occurrences of an Element in a Nested List:

```python
def count_occurrences(lst, elem):
    """
    Counts how many times elem appears in a nested list.

    Parameters:
    lst (list): A nested list of elements.
    elem: The element to count.

    Returns:
    int: Number of times elem appears in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recursive call for sublist
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

nested_list = [[1, 2], [2, 3], [2, 4]]
result = count_occurrences(nested_list, 2)
print("Occurrences of 2:", result)
```

Flatten a List of Lists of Lists:

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list, regardless of depth.

    Parameters:
    lst (list): A nested list of arbitrary depth.

    Returns:
    list: A new flattened list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            # Recursively flatten the sublist
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list

nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)
print("Flattened List:", result)
```

Nested List Average:

```python
def average_nested(lst):
    """
    Calculates the average of all numbers in a nested list.

    Parameters:
    lst (list): A nested list of numbers.

    Returns:
    float: Average of all numbers in the nested list.
    """
    def flatten_and_count(lst):
        """
        Helper function that returns a tuple (total_sum, total_count)
        of all numbers in the nested list.
        """
        total_sum = 0
        total_count = 0
        for item in lst:
            if isinstance(item, list):
                sub_sum, sub_count = flatten_and_count(item)
                total_sum += sub_sum
                total_count += sub_count
            else:
                total_sum += item
                total_count += 1
        return total_sum, total_count

    total, count = flatten_and_count(lst)
    if count == 0:
        return 0  # avoid division by zero
    return total / count

nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)
print("Average:", result)
```
